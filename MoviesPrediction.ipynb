```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
data_set=pd.read_csv('/content/imdb_movies.csv')
```

```python
data_set.head()
```

```python
data_set.shape
```

```python
data_set['index']=data_set.index
data_set.head(3)
```

```python
data_set.columns
```

```python
selected_features=['names','overview','genre','crew']
```

```python
for feature in selected_features:
  data_set[feature]=data_set[feature].fillna('')
```

```python
Combined_features=data_set['names']+' '+data_set['overview']+' '+data_set['genre']+' '+data_set['crew']
```

```python
vectorizer=TfidfVectorizer()
```

```python
feature_vectors=vectorizer.fit_transform(Combined_features)
```

```python
similarity=cosine_similarity(feature_vectors)
similarity.shape
```

```python
def predictMovies(movie_name):

    list_all_titles=data_set['names'].tolist()

    movie_index=difflib.get_close_matches(movie_name,list_all_titles)

    close_match=movie_index[0]

    movie_index=data_set[data_set['names']==close_match]['index'].values[0]

    similarity_score = list(enumerate(similarity[movie_index]))

    sorted_list=sorted(similarity_score,key=lambda x:x[1],reverse=True)

    for i in range(20):
      index=sorted_list[i][0]
      print(data_set.iloc[index]['names'])

predictMovies("erotic ghost strory")
```
